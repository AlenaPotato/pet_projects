## A Tree

```python
# I create the tree usig Koch snowflake example
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/fill.html#sphx-glr-gallery-lines-bars-and-markers-fill-py

def koch_snowflake(order, scale=10):
    def _koch_snowflake_complex(order):
        if order == 0:
            # initial triangle
            angles = np.array([0, 120, 240]) + 90
            return scale / np.sqrt(3) * np.exp(np.deg2rad(angles) * 1j)
        else:
            ZR = 0.5 - 0.5j * np.sqrt(3) / 3

            p1 = _koch_snowflake_complex(order - 1)  # start points
            p2 = np.roll(p1, shift=-1)  # end points
            dp = p2 - p1  # connection vectors

            new_points = np.empty(len(p1) * 4, dtype=np.complex128)
            new_points[::4] = p1
            new_points[1::4] = p1 + dp / 3
            new_points[2::4] = p1 + dp * ZR
            new_points[3::4] = p1 + dp / 3 * 2
            return new_points

    points = _koch_snowflake_complex(order)
    x, y = points.real, points.imag
    return x, y
```

```python
# Drawing the trunk
plt.figure(figsize=(8, 8))

y_trunk = []
x_trunk = range(-4,5)
for n in x_trunk:
  if n <=0:
    y=abs(n/2)
    y_trunk.append(y)
  else:
    y = n/2
    y_trunk.append(y)

y_trunk_rev = []
for n in x_trunk:
  if n <=0:
    y=-abs(n/4)
    y_trunk_rev.append(y)
  else:
    y = -n/4
    y_trunk_rev.append(y)

plt.plot(x_trunk, y_trunk, lw=12, color='r', alpha=0.9)
plt.plot(x_trunk, y_trunk, lw=12, color='g', alpha=0.55)
plt.plot([0,0], [0,4], lw=12, color='r', alpha=0.9)
plt.plot([0,0], [0,4], lw=12, color='g', alpha=0.55)


plt.plot(x_trunk, y_trunk_rev, lw=12, color='r', alpha=0.9)
plt.plot(x_trunk, y_trunk_rev, lw=12, color='g', alpha=0.55)


# Creating leaves
x, y = koch_snowflake(order=2)

import random

rand_list=[]
n=48
random.seed(30983265)
for i in range(n):
    rand_list.append(random.randint(30,150))

for i in range(0, len(rand_list)):
  if y[i] < -2.6:
    plt.plot(x[i], y[i], 'o', ms=rand_list[i], c='w', alpha=0.3)
  else:
    plt.plot(x[i], y[i], 'o', ms=rand_list[i], c='g', alpha=0.8)

# And a little stump
plt.plot([0,0], [-6,0], lw=22, color='r', alpha=0.9)
plt.plot([0,0], [-6,0], lw=22, color='g', alpha=0.55)

plt.ylim(-6, 7)
plt.show()
```
